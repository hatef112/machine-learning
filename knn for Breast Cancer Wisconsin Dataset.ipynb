# import libaries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import preprocessing
%matplotlib inline
```

# read dataset

```python
df = pd.read_csv('Breast Cancer Wisconsin Dataset.csv')
df.head()
```

# delete some columns

```python
df = df.drop(columns='Unnamed: 32')
df = df.drop(columns='id')
```

```python
df.head()
```

```python
print(df.columns)
```

# define x & y

```python
x = df[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']].values
y = df[['diagnosis']].values
```

# scale x for better performence

```python
x = preprocessing.StandardScaler().fit(x).transform(x.astype(float))
```

# split train & test

```python
from sklearn.model_selection import train_test_split
y = np.where(df["diagnosis"] == "M", 1, 0)
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=4)
```

# define knn model

```python
from sklearn.neighbors import KNeighborsClassifier
```

# fit the model

```python
k = 9
my_model = KNeighborsClassifier(n_neighbors = k).fit(x_train, y_train.ravel())
```

# predict results

```python
yhat = my_model.predict(x_test)
```

# model score

```python
from sklearn import metrics
from sklearn.metrics import f1_score
print("Train set Accuracy: ", metrics.accuracy_score(y_train, my_model.predict(x_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
f1 = f1_score(y_test, yhat)
print("F1 Score:", f1)
```
